```python
first_name='adv'
last_name='lovelace'
full_name=f'{first_name}+{last_name}'
#要不要这个+号？在f字符串里，不能有。直接空格。
full_name=f'{first_name} {last_name}'
#f字符串的应用。不一定在括号内。f:formate的简写
print(full_name)
```

```python
first_name='adv'
last_name='lovelace'
full_name=first_name+last_name
#这是字符串拼接
# full_name={first_name}+{last_name}
# 错误1：full_name={first_name}+{last_name} 语法错误
# 这行代码不能这样写。
# 花括号 {} 只有在 f-string（f'...'） 里面才表示“插入变量”，
# 在普通赋值语句里，它们没有特殊含义，会被 Python 当成一个集合（set）或字典（dict）来解析，从而导致语法错误。

full_name=f'{first_name} {last_name}'
print(full_name)
message=f'hello,{full_name.title()}'
print(message)
```

```python
# mm="language\npython\nC\njavascript".upper()
# print(mm)

print('language\npython\nc\njavascript'.upper())
#---->LANGUAGE
#---->PYTHON
#---->C
#---->JAVASCRIPT

print("'language\npython\nc\njavascript'.upper()")
#---->'language
# ---->python
# ---->c
# ---->javascript'.upper()   python把双引号里面的内容当作一个字符串,但是也执行了引号内的\n换行

print("languages:\n\tpython\n\tc\n\tjavascript".upper())

```

```python
favorite_language='python '
# print(favorite_language)

favorite_language.rstrip()
#--->'python'

favorite_language
#--->'python '这时候重新访问favorite_language，结果并没有改变。因为刚才的rstript()后，并没有重新赋值给原来的变量名favorite_language
#要想更改原字符串，就要重新赋值：
favorite_language=favorite_language.rstrip()
print(favorite_language)
#此处看不出效果，输出没有引号，看不出右侧空白已经删除。用terminal执行可看出。
```

```python
favorite_language=" python "
print(favorite_language.lstrip())
#---->'python '
favorite_language
print(favorite_language.strip())
#--->'python'
```

```python
s='http://www.python.org'
d=s.removeprefix('http://')
print(s)
#--->http://www.python.org 没有变化，原来的变量值没变
print(d)
#--->www.python.org   修改后重新赋值给新的变量。
```
