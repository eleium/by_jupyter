```python
# 用for 循环 遍历整个列表：
magicians = ["alice", "david", "carolina"]
for magician in magicians:
    print(magician)

# 重复执行 print(magician),每次在列表magicians中按顺序取一个元素，有几个元素，重复几遍。
```

```python
# 在for 循环中，可以对每个元素执行任意操作：
magicians = ["alice", "david", "carolina"]
for magician in magicians:
    print(f"{magician}的表演太精彩了，what a great trick!")
    # 对元素进行操作，增加首字母大写。并让该行与下一行之间空一行。
    print(f"tell {magician.title()} please,I like his show.\n")

# 跳出循环，这一行是主代码了。并且只执行一次：
print("Thank you,everyone.")
```

```python
# 缩进错误导致的逻辑错误示例：
magicians = ["alice", "david", "carolina"]
for magician in magicians:
    print(f"{magician.upper()}\tthat was g great trick!")
# 原本应该在缩进内的，放到缩进外，只能循环了一次，产生了逻辑错误：
print(f"\nI can't wait to see your nexe trick,{magician.title()}.")
```

```python
# 创建一个数列，包含前10个整数的平方：
for i in range(1, 11):
    i = i**2
    # print(i)

    # 这种写法，只是打印了所有的1到10的平方值。而且是逐行打印
    # 想要同行打印，需要print(i,end=" "),把print()默认的里面有的\n（换行）变成 空格。
    # print()函数默认有一个\n转义符，作用是： 另起一行。
    print(i,end=' ')
```

```python
# 要把这个数列放到一个列表里，需要用到append()方法：
# 先创建一个空列表，然后每次得到一个结果，就append()增加到这列表里：
num = []
for i in range(1, 11):
    # 把每一次的取的数，取平方计算，然后赋值给一个变量名
    square = i = i**2  # 这种写法等同于有square = i**2
    # 把这个变量增加到num空列表里
    num.append(square)
print(num)
#得到一个新的列表。
```

```python
#简化方式：
num=[]
for i in range(1,11):
    #去掉了中间的临时变量： square=i**2,在append()方法里直接用了 i**2
    num.append(i**2)
print(num)
#这种方法更简便，但是没有上面的写法逻辑清晰易懂。推荐上面的写法。代码的可读性。
```

```python
#立方：有三种写法
#1，用临时的变量cube
list_5=[]

for i in range(1,11):
    cube=i**3
    list_5.append(cube)
print(list_5)
    
#2,直接用表达式：i**3
list_5=[]
for i in range(1,11):
    i=i**3
    list_5.append(i)
print(list_5)

#3,极致简化，把中间的i=i**3放到append(i**3)当参数
list_5=[]
for i in range(1,11):
    list_5.append(i**3)
    print(list_5) #循环内，打印出矩阵

#以上三种，第一要注意print()要在循环之外。否则打印的是一个矩阵。
#第二，不能出现二次立方的循环：i=i**3,然后append(i**3)。循环里重复计算了两次 i**3（一次赋值给 i，一次 append）
```

```python
#切片slice :取列表里的索引下标的一段 
player=['charles','martina','michael','florence','eli']
print(player[0:3])
#[0:3]表示取0-3（不含3）的下标元素，即：0，1，2.
#注意不能用[0,3]

print(player[1:4])
```

```python
#切片slice的截取  list[start:stop:step]
player=['charles','martina','michael','florence','eli']
#如果没有指定下标的开始位置，默认从头（0）开始
print(player[:3])

#如果没有指定下标的结束位置，默认到结尾：
print(player[2:])

#负数下标也可以使用，就是从后往前倒数：
print(player[-2:],'\n')
#取后面两个

#对比方括号里面冒号的位置：
print(player[:-2])
#表示：从0到-2之间的元素，即0，1，2

print(player[-2]) 
#获取的是倒数第二个下标的元素值

#步进：
print(player[0:4:2])
#--->['charles','michael'] 0：charles,隔两个，第三个3：‘michael'.再隔两个，第六个，列表里没有。
print(player[::2])
#--->['charles', 'michael', 'eli'],其实是隔一个。

print(player[2::])
#表示从2开始到最后，步长为1.等同与 player[2:],即：player[2:stop:1]
```

```python
# ✅ 所有可能的写法（以 player 为例）
# python
# player = ['charles', 'martina', 'michael', 'florence', 'eli']
# 写法	                     含义	                         结果
# player[:]	                 完整复制	                     全部元素
# player[::]	             同上	                         全部元素
# player[0:]	             从开头到末尾	                 全部元素
# player[:len(player)]	     到末尾	                         全部元素
# player[0:len(player):1]	 全参数完整写法	                 全部元素
# player[0::1]	             开头到末尾，步长 1             全部元素
# player[:len(player):1]	 同上	                         全部元素
# player[::1]	             同上	                         全部元素
# player[...]	❌ 没有这种写法	—
```

```python
#复制列表：用[:]即可：  注意，复制后得到的是原列表的一个副本。后续操作与原列表无关。
player=['charles','martina','michael','florence','eli']
player_new_1=player[:]

print(f'player是老的团队,队员是{player}')
print(f'\nplayer_new_1是新的团队,队员是{player_new_1}')
```

```python
#得到副本后的操作，与原列表无关。变成了两个独立的列表。
player=['charles','martina','michael','florence','eli']
player_new_1=player[:]
player.append('jhon')
player_new_1.append('harry')

print(f'新添加一人。队员名单是{player}')
print(f'新添加一人，队员名单是{player_new_1}')
```

```python
#必须用new_list=list[:],才能这样分别操作新旧两个列表，如果用new_list=list,是不行的：
#这就是切片操作的优势：得到一个副本。

player=['charles','martina','michael','florence','eli']
player_new_1=player

player.append('jhon')
player_new_1.append('harry')

print(f'新添加一人。队员名单是{player}')
print(f'新添加一人，队员名单是{player_new_1}')

#player 先执行append('jhon')--->['charles', 'martina', 'michael', 'florence', 'eli', 'jhon']
#然后再执行： player_new_1.append('harry')------>['charles', 'martina', 'michael', 'florence', 'eli', 'jhon', 'harry']

#因为append()方法会永久改变原列表，所以原来的列表就变成了最后的结果。

#根本原因是：player_new-1=player  这样就把一个列表，起了两个名字。列表还是原来的那一个列表。
```
