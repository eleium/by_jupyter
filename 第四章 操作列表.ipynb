```python
# 用for 循环 遍历整个列表：
magicians = ["alice", "david", "carolina"]
for magician in magicians:
    print(magician)

# 重复执行 print(magician),每次在列表magicians中按顺序取一个元素，有几个元素，重复几遍。
```

```python
# 在for 循环中，可以对每个元素执行任意操作：
magicians = ["alice", "david", "carolina"]
for magician in magicians:
    print(f"{magician}的表演太精彩了，what a great trick!")
    # 对元素进行操作，增加首字母大写。并让该行与下一行之间空一行。
    print(f"tell {magician.title()} please,I like his show.\n")

# 跳出循环，这一行是主代码了。并且只执行一次：
print("Thank you,everyone.")
```

```python
# 缩进错误导致的逻辑错误示例：
magicians = ["alice", "david", "carolina"]
for magician in magicians:
    print(f"{magician.upper()}\tthat was g great trick!")
# 原本应该在缩进内的，放到缩进外，只能循环了一次，产生了逻辑错误：
print(f"I can't wait to see your nexe trick,{magician.title()}.\n")
```

```python
# 不该缩进而缩进了：
magicians = ["alice", "david", "carolina"]
for magicain in magicians:
    print(f"{magicain.title()},I like your show!")
    # 下面的不应该放到循环体内，造成逻辑错误。被for循环看作一体，一起重复执行。
    print("enjoy tonight,everyone!\n")
```

```python
# 创建数值列表：range()函数。注意，不包含最后一个数。

for value in range(1, 5):
    print(value)
    # python中的差一行为。不执行5，只执行1，2，3，4.
```

```python
# 如果range()只有一个参数，那么默认从0-n-1:
for i in range(5):
    print(i)
```

```python
# 创建数值列表： 用list()函数，创建range()函数的列表。
# 说明一个函数可以当作另外一个函数的参数：
list1 = list(range(5))
# 创建0，1，2，3，4列表
print(list1)
```

```python
# range()的第三个参数：step 步长
# 用第三个参数创建偶数列表：
list2 = list(range(0, 10, 2))
# list2=list(range(10,2)) 这种写法不对。python会认为是递减数列，并且步长是1，方向错误。
print(list2)
```

```python
# 创建一个数列，包含前0个整数的平方：
for i in range(1, 11):
    i = i**2
    print(i)

    # 这种写法，只是打印了所有的1到10的平方值。而且是逐行打印
    # 想要同行打印，需要print(i,end=" "),把print()默认的里面有的\n（换行）变成 空格。
    # print()函数默认有一个\n转义符，作用是： 另起一行。
```

```python
# 要把这个数列放到一个列表里，需要用到append()方法：
# 先创建一个空列表，然后每次得到一个结果，就append()增加到这列表里：
num = []
for i in range(1, 11):
    # 把每一次的取的数，取平方计算，然后赋值给一个变量名
    square = i = i**2  # 这种写法等同于有square = i**2
    # 把这个变量增加到num空列表里
    num.append(square)
print(num)
```

```python
#简化方式：
num=[]
for i in range(1,11):
    #去掉了中间的临时变量： square=i**2,在append()方法里直接用了 i**2
    num.append(i**2)
print(num)
#这种方法更简便，但是没有上面的写法逻辑清晰易懂。推荐上面的写法。代码的可读性。
```

```python
#简单的数值列表计算： min()求最小值 , max()求最大值 , sum()函数求求和.
digits=[1,2,3,4,5,6,7,8,9,0]
min(digits)
max(digits)
sum(digits)
#jupyter 默认在没有print()显式打印的情况下，默认值显示最后一行的结果。
```

```python
#列表推导式：
num_list=[i**2 for i in range(1,11) ]
#格式： [表达式 for I in 可迭代对象list if(条件过滤，可选)]      直接对该列表(列表是可迭代对象)操作。
print(num_list)
```

```python
# 更新列表中每个元素的值为自身的2倍
list1 = [1, 2, 3, 4, 5]
for i in range(len(list1)):  
    # ---->len(list1)列表长度5---->range(5)创建0-4的数列---->for I in range(5)---->循环5次：i从0-4
    list1[i] = list1[i] * 2  
    # 这一行一般叫：循环体。for I in range()叫循环条件，也叫循环头。
    #这一行实现了列表中的元素值的更改： list[index] = 'new_value'
print(list1)

# [表达式 for I in 可迭代对象list if(条件过滤，可选)]      直接对该列表(列表是可迭代对象)操作。
```
