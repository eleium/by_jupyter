```python
name = ["jack", "rose", "rose", "bob", "white"]
# 列表list的 增  append insert
name.append("lucy")
# append:在列表的最后，增加新的元素
print(name)
# ---->['jack','rose','rose','bob','white','lucy']

name.insert(2, "jimy")
# 在第2下标处增加新元素，而 非 第二下标后面。
print(name)
# --->['jack','rose','jimy','rose','bob','white','lucy']
# insert: 随意在任何位置增加新元素


print(name.remove("rose"))
# list的 删 del语句  remove()方法针对元素值  pop()方法针对元素下标
# ---->None  因为remove()只是修改了列表，但是没有返回值
print(name)
# --->['jack','jimy','rose','bob','white','lucy']
# remove: 删除特定的元素：指定的字符串.删除列表中的第一个符合的元素
# remove()方法 知道元素的值，不需要知道元素的index下标。

# 用remove()方法删除了其中的一个元素后，也跟pop()方法一样，可以继续使用这个被删除（提取）的元素：
# ////这句话是错的（但你后面又自己纠正了）。remove() 不能像 pop() 那样直接返回被删元素，只能提前赋值。
first_friend = name.remove("rose")
# 将被删除的元素赋值给一个新的变量名，就可以继续使用了
print(f"my first friend is {first_friend}")
# -->my first friend is None
# 因为 .remove() 方法没有返回值，它返回的是 None。
# 你把被删除的元素赋值给 first_friend，其实赋的是 None，而不是被删除的那个元素。
# .remove() 的作用是直接修改原列表，找到第一个匹配的元素并删除它。
# 它不返回被删除的元素，只返回 None（和 .append()、.sort() 等原地修改方法一样）。

# 要继续使用被remwove()删除的元素，先要把这个元素赋值给一个变量：
first_friend_of_mine = "rose"
# 一定要把这个将要被删除的元素先赋值给一个变量。
print(f"我交的第一个朋友是{first_friend_of_mine}")
# ---->我交的第一个朋友是rose


del name[3]
print(name)
# --->['jack',‘jimy','rose','white','lucy']
# 注意： 这是del 语句，而非方法，所以del放到list前面，并且没有 . ，直接是删除列表中的第[n]元素。


# 注意：对列表的操作完成后，就是改变了原列表。这一点与字符串完全不同。
# 因为列表可变，字符串不可变。

name[1] = "merry"
print(name)
# --->['jack','merry','rose','white','lucy']
# 列表的 改  直接赋值原来的下标元素值为新的元素值
```

```python
fruit = ["apple", "orrange", "banana"]
poped_fruit = fruit.pop()
print(fruit)
# --->['apple','orrange']  没有了 'banan',最后一个元素已经被弹走了。
print(poped_fruit)
# ---->被弹走（提取）的元素。可以放到别的地方去应用。
# pop()也算一种列表的删除的办法吧。：提取最后一个元素，并保存这个被提取的元素。

# 如果用pop(index)提取列表中的任意一个元素：把要提取的元素的下标添加进pop()里。
new_popped_fruit = fruit.pop(0)
print(fruit)
# --->['orrange']
print(f"我最喜欢吃的是{new_popped_fruit.upper()}!")
# ---->我最喜欢吃的是APPLE!


# 如果删除以后不需要了，就用del语句。如果提取走了，还要用，就用pop(index).空的默认最后一个元素。
```

```python
# 列表的排序 sort()方法和sorted()函数
cars = ["bmw", "audi", "toyota", "subaru"]

# sort:按照字母的先后顺序排序，并永久改变原列表的元素顺序
cars.sort()
print(cars)
# --->['audi','bmw','subaru','toyota']

# 如果想得到相反的排序，就用参数： revese=True
cars = ["bmw", "audi", "toyota", "subaru"]
cars.sort(reverse=True)
# sort是对字母顺序排列，用了reverse=True也一样，是倒数字母，而不是把原列表顺序相反。
print(cars, "\n")
# ---->['toyota', 'subaru', 'bmw', 'audi']

# 如果想临时的排序，用函数sorted()，从而保留原来的列表元素顺序。

# 函数(obj)和obj.方法()的使用方法不同，要注意。

cars = ["toyota", "subaru", "bmw", "audi"]
print(sorted(cars))
# ---->['audi', 'bmw', 'subaru', 'toyota']  按照字母顺序排列
print(cars)
# ---->['toyota', 'subaru', 'bmw', 'audi'] 原来的顺序没变

# 同样的，可以给函数sorted()传参：reverse=True,实现列表的字母顺序的倒序。
print(sorted(cars, reverse=True))
# ---->['toyota', 'subaru', 'bmw', 'audi']
```

```python
# 直接反转列表里的元素顺序，而不是按照字母顺序排列，用reverse()方法。
cars = ["bmw", "audi", "toyota", "subaru"]
cars.reverse()
# 注意：不能直接print(cars.reverse())，这样得到的是None
# 方法(method) 永久改变原序列，并且没有返回值，或者说返回值是None.所以不要用print()函数打印，否则打印结果是：None
print(cars)
# ---->['subaru','toyota','audi','bmw']
```

```python
# 用len()函数确定列表的长度：
cars = ["bmw", "audi", "toyota", "subaru"]
len(cars)
print(len(cars))
# ---> 4  因为len()是函数，所以有返回值，所以就可以用print()函数直接打印。
```
