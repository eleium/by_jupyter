````python
# 任意多个参数： *args,和任意多的关键字参数:*kwargs
def decribe_pet(*name, **types):
    # *name 会把所有传入的实参打包成一个元组 ('糖宝', '猫')。
    print(f"我喜欢{name},因为我喜欢")


decribe_pet("糖宝", types="猫", types_su="little cat")
# 多个实参，不管是位置参数，还是关键字参数，变量名不能重复。

# 因为有形参，所以必须有实参。
decribe_pet("糖宝", "猫", "花花")
# 实参和形参的数量必须一致。

# ✅ 什么时候“必须一致”？
# 函数定义	实参数量要求	示例
# 普通形参 def f(a, b):	✅ 必须一致	f(1, 2) ✅ / f(1) ❌ / f(1, 2, 3) ❌
# 默认参数 def f(a, b=2):	⚠️ 可以少传	f(1) ✅ / f(1, 3) ✅
# 可变参数 def f(*args):	❌ 任意数量	f() ✅ / f(1) ✅ / f(1, 2, 3) ✅
# 关键字参数 def f(**kwargs):	❌ 任意数量	f() ✅ / f(a=1) ✅ / f(a=1, b=2) ✅
# 🔍 示例


# 1. 普通形参：必须一致
def f1(a, b):
    print(a, b)


f1(1, 2)  # ✅
# f1(1)         # ❌ TypeError


# 2. 默认参数：可以少传
def f2(a, b=2):
    print(a, b)


f2(1)  # ✅ 1 2
f2(1, 3)  # ✅ 1 3   这是把b=2的默认参数，被b=3的实参给代替了。


# 3. 可变参数：任意数量
def f3(*args):
    print(args)


f3()  # ✅ ()
f3(1)  # ✅ (1,)
f3(1, 2, 3)  # ✅ (1, 2, 3)


# 4. 关键字参数：任意数量
def f4(**kwargs):
    print(kwargs)


f4()  # ✅ {}
f4(a=1)  # ✅ {'a': 1}
f4(a=1, b=2)  # ✅ {'a': 1, 'b': 2}


# 📌 一句话总结
# “实参和形参数量必须一致”只对普通形参成立。有默认参数、可变参数 *args 或关键字参数 **kwargs 时，数量可以不一致。
````

````python
# 📌 语法规则
# 调用方式	                写法	                                         说明
# 多个关键字参数	    func(key1=value1, key2=value2, key3=value3)	        用 = 连接，逗号分隔
# 传入字典解包	        func(**{"key1": value1, "key2": value2})	        用 ** 解包字典


# ✅ 两种传递方式对比

# # 方式1：直接写多个关键字参数
# show_info(name="张三", age=18, city="北京")

# # 方式2：先用字典存，再解包
info = {"name": "张三", "age": 18, "city": "北京"}
# show_info(**info)   # ← 注意 **
# 解包是用 **，把字典里面的键值对取出来，作为关键字参数传入函数。
# 因为字典是键值对，一对有俩元素，所以用两个**。

# 📌 * 和 ** 解包对比
# 符号	解包对象	           传入方式	           示例
# *        	列表/元组	       位置参数	          func(*[1, 2, 3]) → func(1, 2, 3)
# **	    字典	           关键字参数	      func(**{"a": 1, "b": 2}) → func(a=1, b=2)


# 📌 完整示例（含位置参数 + 关键字参数）


def show_info(name, **kwargs):
    print(f"姓名：{name}")
    for key, value in kwargs.items():
        print(f"{key}：{value}")


# show_info("张三", age=18, city="北京", hobby="编程")

# 输出：
# 姓名：张三
# age：18
# city：北京
# hobby：编程


# 📌 一句话总结
# 调用时写多个关键字参数，用逗号分隔：func(key1=value1, key2=value2, ...)。也可以用字典解包：func(**dict)。
````

````python
## 练习8.7 专辑：
# 编写一个 make_album()函数，它创建一个描述音乐专辑的字典。
# 函数接受歌手名和专辑名，并返回一个包含这两项信息的字典。
# 使用这个函数创建三个表示不同专辑的字典，并打印每个返回值。以核实字典正确的储存了专辑的信息。

# 给make_album()添加一个默认值为None的可选形参，以便储存专辑包含的歌曲数量。
# 如果调用的时候指定了歌曲数量，就把这个值添加到字典中。
# 调用这个函数，并至少一次调用指定专辑包含的歌曲数量。


def make_album(singer, cd_name, num=None):
    # 形参的num=""和num=None效果相同。但是None更清晰。如果用""，而用户输入的是0，""代表空字符串，0不是。
    if num:
        # if num 不是“是否传递了参数”，而是“传递进来的值是否非空（非零、非 None、非空字符串）”。
        # 你传递了 num=5，所以 num 是 5，if num 为 True。
        dict_cd = {singer: cd_name, "songs": num}
    else:
        dict_cd = {singer: cd_name}

    return dict_cd


print(make_album("谭咏麟", "爱情陷阱", num=5))
print(make_album("张国荣", "沉默是金"))
print(make_album("许巍", "故乡"))
````

````python
# 练习8.8用户的专辑
# 在8.7中，编写一个while循环。让用户输入歌手名和专辑名。
# 获得这些信息后，使用它们来调用make_album()函数并将创建的字典打印出来。
# 循环务必有退出途径。

## 练习8.7 专辑：
# 编写一个 make_album()函数，它创建一个描述音乐专辑的字典。
# 函数接受歌手名和专辑名，并返回一个包含这两项信息的字典。
# 使用这个函数创建三个表示不同专辑的字典，并打印每个返回值。以核实字典正确的储存了专辑的信息。

# 给make_album()添加一个默认值为None的可选形参，以便储存专辑包含的歌曲数量。
# 如果调用的时候指定了歌曲数量，就把这个值添加到字典中。
# 调用这个函数，并至少一次调用指定专辑包含的歌曲数量。


def make_album(singer, cd_name, num=None):
    if num:
        dict_cd = {singer: cd_name, "songs": num}
    else:
        dict_cd = {singer: cd_name}

    return dict_cd


while True:
    singer = input("请输入 歌手名： ")
    if singer == "quit":
        break
    cd_name = input("请输入专辑名")
    if cd_name == "quit":
        break
    songs = input("输入歌曲数量")
    if songs:
        album = make_album(singer, cd_name, songs)
    else:
        album = make_album(singer, cd_name)

    print(album)
# 这个print(album)要么放到循环内，要么在循环外先定义。这样写会报错。
````

````python
"""'、'.join(toppings)` 用法精讲

你遇到它好几次了，今天彻底搞明白它。其实结构非常固定，记住一次就能一直用。

 🧩 一、固定的结构（记这一个就够了）

"分隔符".join(可迭代对象)


这永远是一个**字符串方法**，由三部分组成：

| 部分 | 含义 | 示例 |

| `"分隔符"` | 你想用什么符号连接 | `"、"`、`", "`、`"-"`、`"\n"` |
| `.join()` | 固定的方法名 | 没有任何变化 |
| `(可迭代对象)` | 列表、元组、字符串等 | `["桔子", "葡萄"]` |



 🧪 二、看几个例子（比文字好懂）


# 1. 用顿号连接
"、".join(["桔子", "葡萄", "鸭梨"])
# → "桔子、葡萄、鸭梨"

# 2. 用逗号+空格连接
", ".join(["apple", "banana", "cherry"])
# → "apple, banana, cherry"

# 3. 用空字符串连接
"".join(["a", "b", "c"])
# → "abc"

# 4. 用换行符连接
"\n".join(["第一行", "第二行", "第三行"])
# → "第一行\n第二行\n第三行"


 🔍 三、为什么不用 `+` 拼接？


 ❌ 不推荐：手动拼接
result = ""
for item in list:
    result += item + "、"
# → "桔子、葡萄、鸭梨、"（注意末尾多了顿号）

 ✅ 推荐：join
"、".join(list)
# → "桔子、葡萄、鸭梨"（干净，没有多余符号）
```

**`join` 的优势：**

| 对比                 | `+` 拼接                 | `.join()` |

| 末尾多余分隔符       | ❌ 容易出现              | ✅ 不会出现 |
| 性能                 | ❌ 多次拼接（慢）        | ✅ 一次处理（快） |
| 可读性               | ⚠️ 代码长               | ✅ 一行搞定 |

---

 📌 四、什么时候用？

| 场景                           | 是否适合 `join` |

| 把列表内容拼成一句话          | ✅ 非常适合 |
| 把列表内容拼成 CSV 行         | ✅ 非常适合 |
| 打印所有参数                  | ✅ 非常适合 |
| 拼接单个字符串                | ❌ 不需要，直接用 `"" + ""` |
| 可迭代对象是数字              | ⚠️ 要先用 `map(str, list)` 转成字符串 |

---

⚠️ 五、常见陷阱


 ❌ 错误：数字不能直接用 join
", ".join([1, 2, 3])
 TypeError: sequence item 0: expected str instance, int found

 ✅ 正确：先转成字符串
", ".join(map(str, [1, 2, 3]))
# → "1, 2, 3"



 📌 一句话总结

 "分隔符".join(可迭代对象)` 是 Python 中拼接字符串的标准方法。
 它把列表/元组中的每个元素用指定分隔符连接成一个字符串，简洁、高效、不会有多余符号。


    转换函数str()和内置函数map()的区别：

 **你问到了一个非常关键的区别！** `str()` 和 `map(str, list)` 完全不同，它们解决的不是同一个层面的问题。



 ✅ 直接回答

| 用法 | 作用 | 返回值 |

| `str(123)`                      | 把**单个**数字转成**单个**字符串                           | `"123"` |
| `map(str, [1, 2, 3])`          | 把**每个**元素都转成字符串，并返回一个**迭代器**            | `<map object>` |

---

🔍 为什么需要 `map`？

`.join()` 要求列表中的**每个元素都是字符串**：


", ".join([1, 2, 3])   # ❌ TypeError: sequence item 0: expected str instance, int found


你不能直接把数字列表传给 `join`，必须先把所有数字变成字符串。

 🧪 对比三种写法

1. 手动逐个转换（最笨）
items = [1, 2, 3]
str_items = []
for i in items:
    str_items.append(str(i))
print(", ".join(str_items))
# → "1, 2, 3"

 2. 列表推导式（常见的写法）
items = [1, 2, 3]
print(", ".join([str(i) for i in items]))
# → "1, 2, 3"

 3. map（更简洁）
items = [1, 2, 3]
print(", ".join(map(str, items)))
# → "1, 2, 3"


 📌 `map()` 是什么？


map(str, [1, 2, 3])

- `map` 是 Python 内置函数
- 它把 `str` 这个函数**应用到列表的每个元素上**
- 返回一个**迭代器**（不会一次性生成所有结果，节省内存）
- `.join()` 会遍历这个迭代器，把所有字符串连接起来



 📌 什么时候用哪个？

| 场景 | 推荐写法 |
| :--- | :--- |
| 只转一个数字                            | `str(123)` |
| 把列表里所有数字转成字符串              | `map(str, list)` 或 `[str(i) for i in list]` |
| 需要进一步处理每个元素                  | 列表推导式 |
| 只用来传给 `.join()`                    | `map(str, list)` 更简洁 |

---

📌 一句话总结

`str(123)` 只转一个；
`map(str, [1,2,3])` 把每个元素都转一次。
`.join()` 需要所有元素都是字符串，所以先用 `map(str, list)` 把整个列表转成字符串，再用 `join` 拼接。

"""
````

````python
# 导入特定的函数：
# from module_name import function_name


# 如果要从一个模块中导入多个函数，可以用逗号分开：
# from module_name import function_0,function_1,function_2

# 前面的例子，可以这样写：省去了再写一遍模块名的过程，因为已经显式的导入具体的模块的函数了。
from pizza import make_pizza

make_pizza(16, "pepperoni")
make_pizza(12, "mushroom", "green peppers", "extra cheese")
````
