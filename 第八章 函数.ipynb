```python
# 函数的形参和实参可以多个。
# 向函数传递实参的方式：位置参数*args，或者关键字参数*kwargs:每个实参以变量名和值组成。也可以是列表和字典。

# 位置参数：

# 显示宠物信息，指出 一个宠物属于哪种动物，以及它们叫什么名字：


def message_animal(type, name):
    # 表示这个函数需要传进来两个参数，一个是类型的，一个是名字的
    print(f"{name}是一种{type}")


message_animal("猫科", "糖豆")

# 多个位置参数，必须考虑参数的顺序。不需要赋值。
```

```python
# 关键字参数：可以不用考虑形参的顺序了，因为调用的时候，明确了形参= 实参。


def message_animal(name, type):
    print(f"{name} 是 {type}")


message_animal(name="糖宝", type="cat")
# 调用的时候，实参必须把形参赋值，必须用形参=实参的固定格式。


# 🔍 为什么不能用 :？
# 符号	在 Python 中的用途
# :	用于字典（{"name": "糖宝"}）、切片（list[1:3]）、类型注解（def func(x: int):）、语句开头（if、for、def）
# =	用于赋值、关键字参数（func(name="糖宝")）、默认参数（def func(name="糖宝"):）
# 关键字参数的语法就是 key=value，这是 Python 的硬性规定，不能改成 :。
```

```python
# 默认参数：在形参里，先明确指出 参数1='value'


def message_animal(name, type="cat"):
    # 这个type='cat‘就是defoult argumeng,默认参数。
    # 如果不改变默认参数，在实参里可以不写出来。除非要改变默认参数。

    print(f"{name} 是 {type}")


message_animal(name="糖宝")
message_animal(name="糖豆", type="little cat")
# 调用的时候，实参必须把形参赋值，必须用形参=实参的固定格式。

message_animal("糖宝")
# 你的理解完全正确：默认参数可以不写（使用默认值），也可以写（覆盖默认值）。

message_animal("糖宝", type="little cat")

# 调用时可以用位置参数，也可以用关键字参数，但混用时位置参数必须在前。
# 如果你混用位置参数和关键字参数，位置参数必须放在关键字参数前面：


# 🔍 为什么不能用 :？
# 符号	在 Python 中的用途
# :	用于字典（{"name": "糖宝"}）、切片（list[1:3]）、类型注解（def func(x: int):）、语句开头（if、for、def）
# =	用于赋值、关键字参数（func(name="糖宝")）、默认参数（def func(name="糖宝"):）
# 关键字参数的语法就是 key=value，这是 Python 的硬性规定，不能改成 :。
```

```python
# 等效的函数调用：
def describe_pet(name, animal_type="dog"):
    print(f"叫{name}的是{animal_type}")


describe_pet(name="willie")
describe_pet("rose")
describe_pet("rsose", animal_type="little dog")
describe_pet("jace", "rose")

# ⚠️ 唯一规则：位置参数必须在关键字参数之前
# 写法	                                                 是否合法	         说明
# describe_pet("jace", animal_type="rose")	            ✅ 合法	           位置参数在前，关键字参数在后
# describe_pet(name="jace", "rose")	                    ❌ 报错	           关键字参数在前，位置参数在后


# 📌 一句话总结
# 只要顺序不乱（位置参数在前，关键字参数在后），你可以自由组合调用方式。位置参数按顺序匹配，关键字参数按名字匹配。


# 在 Python 中，字符串字面量（文本内容）必须用引号包裹。变量名不用引号。
# 🔍 什么情况必须加引号？
# 情况	                     是否需要引号	                         示例
# 传入字符串内容	        ✅ 必须加	                           describe_pet("rose")
# 传入变量名	            ❌ 不加	                                name = "rose" → describe_pet(name)
# 关键字参数名	            ❌ 不加	                                describe_pet(name="rose")
# 函数名	                ❌ 不加	                                describe_pet(...)
# 类型注解	                ❌ 不加	                                def func(name: str):
```

```python
# 任意多个参数： *args,和任意多的关键字参数:*kwargs
def decribe_pet(*name):
    # *name 会把所有传入的实参打包成一个元组 ('糖宝', '猫')。
    print(f"我喜欢{name},因为我喜欢")


decribe_pet("糖宝", "猫")
# 因为有形参，所以必须有实参。
decribe_pet("糖宝", "猫", "花花")
# 实参和形参的数量必须一致。
```

```python
# 任意多个参数： *args,和任意多的关键字参数:*kwargs
def decribe_pet(*name, **types):
    # *name 会把所有传入的实参打包成一个元组 ('糖宝', '猫')。
    print(f"我喜欢{name},因为我喜欢")


decribe_pet("糖宝", types="猫", types_su="little cat")
# 多个实参，不管是位置参数，还是关键字参数，变量名不能重复。

# 因为有形参，所以必须有实参。
decribe_pet("糖宝", "猫", "花花")
# 实参和形参的数量必须一致。

# ✅ 什么时候“必须一致”？
# 函数定义	实参数量要求	示例
# 普通形参 def f(a, b):	✅ 必须一致	f(1, 2) ✅ / f(1) ❌ / f(1, 2, 3) ❌
# 默认参数 def f(a, b=2):	⚠️ 可以少传	f(1) ✅ / f(1, 3) ✅
# 可变参数 def f(*args):	❌ 任意数量	f() ✅ / f(1) ✅ / f(1, 2, 3) ✅
# 关键字参数 def f(**kwargs):	❌ 任意数量	f() ✅ / f(a=1) ✅ / f(a=1, b=2) ✅
# 🔍 示例


# 1. 普通形参：必须一致
def f1(a, b):
    print(a, b)


f1(1, 2)  # ✅
# f1(1)         # ❌ TypeError


# 2. 默认参数：可以少传
def f2(a, b=2):
    print(a, b)


f2(1)  # ✅ 1 2
f2(1, 3)  # ✅ 1 3   这是把b=2的默认参数，被b=3的实参给代替了。


# 3. 可变参数：任意数量
def f3(*args):
    print(args)


f3()  # ✅ ()
f3(1)  # ✅ (1,)
f3(1, 2, 3)  # ✅ (1, 2, 3)


# 4. 关键字参数：任意数量
def f4(**kwargs):
    print(kwargs)


f4()  # ✅ {}
f4(a=1)  # ✅ {'a': 1}
f4(a=1, b=2)  # ✅ {'a': 1, 'b': 2}


# 📌 一句话总结
# “实参和形参数量必须一致”只对普通形参成立。有默认参数、可变参数 *args 或关键字参数 **kwargs 时，数量可以不一致。
```

```python
# 📌 语法规则
# 调用方式	                写法	                                         说明
# 多个关键字参数	    func(key1=value1, key2=value2, key3=value3)	        用 = 连接，逗号分隔
# 传入字典解包	        func(**{"key1": value1, "key2": value2})	        用 ** 解包字典


# ✅ 两种传递方式对比

# # 方式1：直接写多个关键字参数
# show_info(name="张三", age=18, city="北京")

# # 方式2：先用字典存，再解包
info = {"name": "张三", "age": 18, "city": "北京"}
# show_info(**info)   # ← 注意 **
# 解包是用 **，把字典里面的键值对取出来，作为关键字参数传入函数。
# 因为字典是键值对，一对有俩元素，所以用两个**。

# 📌 * 和 ** 解包对比
# 符号	解包对象	           传入方式	           示例
# *        	列表/元组	       位置参数	          func(*[1, 2, 3]) → func(1, 2, 3)
# **	    字典	           关键字参数	      func(**{"a": 1, "b": 2}) → func(a=1, b=2)


# 📌 完整示例（含位置参数 + 关键字参数）


def show_info(name, **kwargs):
    print(f"姓名：{name}")
    for key, value in kwargs.items():
        print(f"{key}：{value}")


# show_info("张三", age=18, city="北京", hobby="编程")

# 输出：
# 姓名：张三
# age：18
# city：北京
# hobby：编程


# 📌 一句话总结
# 调用时写多个关键字参数，用逗号分隔：func(key1=value1, key2=value2, ...)。也可以用字典解包：func(**dict)。
```
