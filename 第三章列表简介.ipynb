```python
bicycles = ["trek", "redline", "specialized", "cannondale"]
print(bicycles)
print(bicycles[1])
print(bicycles[2].title())
print(bicycles[1])
#
# print(bicycles[0], bicycles[2], bicycles[3])可以一次打印多个指定的元素。
# print(bicycles[0, 2, 3])  和 print(bicycles[0],[2],[3] 这两种都是错的
```

```python
message=f'my first bicycle is {bicycles[0].upper()}'
print(message)
#列表的元素方法，与字符串方法类似。
```

```python
name=['jack','bob','rose','white']
print(name[0],name[2])

print(name[1].title())
# print(name['jack'])   不能直接访问列表里的元素，要通过下标。

print(f'{name[-1].upper()} is 88 years old.')
```

```python
name=['jack','rose','rose','bob','white']
name.append('lucy')
print(name)
#---->['jack','rose','rose','bob','white','lucy']
#列表list的 增

print(name.remove('rose'))
#---->None  因为remove()只是修改了列表，但是没有返回值
print(name)
#--->['jack','rose','bob','white','lucy']
#list的 删

#注意：对列表的操作完成后，就是改变了原列表。这一点与字符串完全不同。
#因为列表可变，字符串不可变。

name[1]='merry'
print(name)
#--->['jack','merry','bob','white','lucy']
#列表的 改
```
