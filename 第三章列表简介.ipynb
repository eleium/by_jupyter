```python
name=['jack','bob','rose','white']
print(name[0],name[2])

print(name[1].title())
# print(name['jack'])   不能直接访问列表里的元素，要通过下标，因为list是有序的可变序列。

print(f'{name[-1].upper()} is 88 years old.')
```

```python
name=['jack','rose','rose','bob','white']
#列表list的 增
name.append('lucy')
#append:在列表的最后，增加新的元素
print(name)
#---->['jack','rose','rose','bob','white','lucy']

name.insert(2,'jimy')
#在第2下标处增加新元素，而 非 第二下标后面。
print(name)
#--->['jack','rose','jimy','rose','bob','white','lucy']
#insert: 随意在任何位置增加新元素


print(name.remove('rose'))
#list的 删
#---->None  因为remove()只是修改了列表，但是没有返回值
print(name)
#--->['jack','jimy','rose','bob','white','lucy']
#remove: 删除特定的元素：指定的字符串.删除列表中的第一个符合的元素

del name[3]
print(name)
#--->['jack',‘jimy','rose','white','lucy']
#注意： 这是del 语句，而非方法，所以del放到list前面，并且没有 . ，直接是删除列表中的第[n]元素。


#注意：对列表的操作完成后，就是改变了原列表。这一点与字符串完全不同。
#因为列表可变，字符串不可变。

name[1]='merry'
print(name)
#--->['jack','merry','rose','white','lucy']
#列表的 改
```

```python
fruit=['apple','orrange','banana']
poped_fruit=fruit.pop()
print(fruit)
#--->['apple','orrange']  没有了 'banan',最后一个元素已经被弹走了。
print(poped_fruit)
#---->被弹走（提取）的元素。可以放到别的地方去应用。
#pop()也算一种列表的删除的办法吧。：提取最后一个元素，并保存这个被提取的元素。

#如果用pop(index)提取列表中的任意一个元素：把要提取的元素的下标添加进pop()里。
new_popped_fruit=fruit.pop(0)
print(fruit)
#--->['orrange']
print(f'我最喜欢吃的是{new_popped_fruit.upper()}!')
#---->我最喜欢吃的是APPLE!
```
