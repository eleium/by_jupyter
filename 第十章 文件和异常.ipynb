```python
# зїГдє†10.5 иЃњеЃҐз∞њ
# зЉЦеЖЩдЄАдЄ™whileеЊ™зОѓпЉМжПРз§ЇзФ®жИЈиЊУеЕ•еРНе≠ЧгАВжФґйЫЖзФ®жИЈиЊУеЕ•зЪДжЙАжЬЙеРНе≠ЧпЉМе∞ЖеЕґеЖЩеЕ•guest_book.txtдЄ≠гАВ
# з°ЃдњЭжЦЗдїґдЄ≠зЪДжѓПдЄАжЭ°иЃ∞ељХйГљзЛђеН†дЄАи°МгАВ
from pathlib import Path

path = Path("guest_book.txt")

names = []

while True:
    name = input("enter your name : ")
    # еЫ†дЄЇи¶Бе§Ъжђ°еЖЩеРНе≠ЧпЉМжЙАдї•inputиѓ≠еП•и¶БеЬ®whileеЊ™зОѓйЗМгАВ
    if name == "quit":
        break
    # жККйААеЗЇжЭ°дїґеЖЩйБУжЬАдЄКжЦєпЉМйБњеЕНдЇЖеЖЩеЬ®еРОйЭҐпЉМquitеЈ≤зїПиҐЂжЙІи°МзЪДзїУжЮЬгАВ

    names.append(name)
    # зЃАеНХзЪДеИЧи°®жЈїеК†еЖЕеЃєгАВ

content = "\n".join(names)
# ж≠§жЧґзЪДnamesжШѓдЄАдЄ™еИЧи°®гАВпЉМзФ®\nеИЖеЙ≤еИЧи°®дЄ≠жѓПдЄАжђ°appendзЪДname.ж≠§жЧґwhileеЊ™зОѓеЈ≤зїУжЭЯгАВ
path.write_text(content)
# жККcontertеЖЩеЕ•жЦЗдїґгАВ

for i in names:
    print(i)
print(type(names))
```

```python
# еИЖжЮРе§ЪдЄ™жЦЗдїґзЪДеНХиѓНжХ∞пЉЪ
# ињЩжЧґпЉМеЫ†дЄЇжШѓйЗНе§НиЃ°зЃЧеНХиѓНзЪДжХ∞е≠ЧпЉМеПѓдї•еЃЪдєЙдЄАдЄ™еЗљжХ∞пЉЪ

path = Path("alice.txt")


def count_words(path):
    """иЃ°зЃЧдЄАдЄ™жЦЗдїґжЬЙе§Ъе∞СеНХиѓН"""
    #еПВжХ∞path=Path('filename'),'filename'жѓПжђ°дїОеИЧи°®дЄ≠жПРеПЦпЉМзђђдЄАжђ°жШѓ'alice'пЉМзђђдЇМжђ°жШѓ "siddhartha.txe",...
    try:
        contents = path.read_text()
        #зїІзї≠жЙІи°М 
    except FileNotFoundError:
        print("зЫЃељХйЗМж≤°жЬЙињЩжЬђдє¶")
    else:
        words = contents.split()
        # wordsиҐЂеЃЪдєЙдЇЖпЉМе∞±и¶БзФ®дЄКгАВ
        num_words = len(words)
        # num_wordsиҐЂеЃЪдєЙдЇЖпЉМе∞±и¶БзФ®дЄКгАВ
        print(num_words)


count_words(path)

filenames = ["alice.txt", "siddhartha.txe", "moby_dick.txt", "little_women.txt"]

for filename in filenames:
    path = Path(filename)
    #жѓПжђ°жККжЦЗдїґзЪДзЫЃељХеѓєи±°еЉХзФ®жЫіжЦ∞гАВ
    count_words(path)
    #и∞ГзФ®еЗљжХ∞гАВ
```
