```python
# 因为 Python 中所有“通过索引或键来获取元素”的操作，统一使用方括号 []。

# 🔍 方括号 [] 的作用
# 数据类型	       用法	                           说明
# 列表	          list[0]	                      通过整数索引获取元素
# 元组	          tuple[0]	                      通过整数索引获取元素
# 字符串	      str[0]	                      通过整数索引获取字符
# 字典	          dict["key"]	                  通过键（key）获取值
# 核心规则： 方括号 [] 是 Python 中“下标访问/键访问”的统一语法，不管底层是序列还是映射。

# . 是用来访问属性和方法的，不是用来访问元素的。

# 有序序列（列表、元组、字符串）有整数下标。
# 字典没有整数下标，但使用键（key）来访问元素。
# 它们都用 []，但括号里的内容不同：一个是数字，一个是键名。

# 集合是无序数列，不能用下标(index)


# 数据类型	      是否有序	       用什么访问	             访问方式
# 列表	         ✅ 有序	         整数索引（下标）	        list[0]
# 元组	         ✅ 有序	         整数索引（下标）	        tuple[0]
# 字符串	     ✅ 有序	         整数索引（下标）	        str[0]
# 字典	         ✅ 有序（3.7+）	 键（key）	                dict["key"]
# 集合	         ❌ 无序	           不支持 []	—
```

```python
# 给字典里  添加  键值对：
alien_0 = {"color": "green", "points": 5}

alien_0["x_position"] = 0
alien_0["y_position"] = 25
print(alien_0)

# #✅ 四种数据结构“增加元素”的方法对比
# 数据类型	                 是否有“增加”操作	              增加方法	                                                   说明
# 列表（list）	                ✅ 可以	                .append(x) .extend([x,y,z])    .insert(index, x)	        末尾追加、合并、插入指定位置
# 元组（tuple）	                ❌ 不可以	              —	                                                       元组是不可变序列，创建后不能修改
# 字典（dict）	                ✅ 可以	                 dict["新键"] = 值   .update({...})	                        添加键值对或合并字典
# 集合（set）	               ✅ 可以	                 .add(x)     .update([x, y])	                            添加单个元素（去重）或合并
```

```python
alien = {}

alien["color"] = "green"
alien["point"] = 5
# 给空字典添加键值对
print(alien)

alien.update({"x_position": 0, "y_position": 25})
# 在原字典里添加另外一个字典.。但是原字典要存在，不能以这种方式创建新字典。
print(alien)
```

```python
# 字典的值的  修改
alien = {"color": "green", "point": 5}

alien["color"] = "red"
print(alien)

# dict["key"] = value 是“添加或修改”的统一操作：键存在就修改，键不存在就添加。它不是两种不同的语法，而是同一个语法的两种结果。

# 修改元素时，字典和列表都是 对象[标识符] = 新值，语法完全一样。不同之处在于新增：字典的键不存在会自动创建，列表的索引超出范围会报错。因为列表有序。
```

```python
alien = {"x_position": 0, "y_position": 25, "speed": "medium"}
print(f"original position:{alien['x_position']}")

# 向右移动外星人：根据当前速度确定将外星人享有移动多远
if alien["speed"] == "slow":
    distance = 1
elif alien["speed"] == "medium":
    distance = 2
elif alien["speed"] == "fast":
    distance = 3

alien["x_position"] = distance + alien["x_position"]
print(f"The new_position is {alien['x_position']}")
```

```python
#用for 循环，同时遍历所有的键和值。
user = {
    "username": "efermi",
    "first_name": "enrico",
    "last_name": "fermi",
}

for key in user:
    print(key)  # 每次循环结果空行
    # 遍历键名,value 同样。这是单独获得所有的键，或者所有的value.

for key, value in user.items():
    print("\n", key, value)
# 同时遍历所有的键和值。
#变量：key 和 value可以是任意的，比如k,v或者name,value等等，前者代表键key，后者代表值value.
#用键和值 循环得到的，是解包。不同于用一个变量，从字典视图得到的是元组。
```

```python
# 字典的方法：
user = {
    "username": "efermi",
    "first_name": "enrico",
    "last_name": "fermi",
}
print(user.keys())
# 获得字典的键

# ---> dict_keys(['username', 'first_name', 'last_name'])
# 不是元组，也不是列表。它是一个 dict_keys 对象，是字典的视图（view）

print(user.values())
# --->dict_values(['efermi', 'enrico', 'fermi'])

print(user.items())
#--->dict_items([('username', 'efermi'), ('first_name', 'enrico'), ('last_name', 'fermi')])
```

```python
for i in user.items():
    print(i)

#直接得到键值对
#从字典视图得到的，是元组。一下子得到一对。
```

```python
# 按照特定的字母顺序来访问字典：
user = {
    "username": "efermi",
    "first_name": "enrico",
    "last_name": "fermi",
}

for i in sorted(user.keys()):
    # 用sorted()函数，以字母顺序从前往后遍历字典的key
    print(f"{i.title()},thank you for take the poll.")


# for i in user.keys().sort():
#     print(i)
# 报错
# 因为 user.keys() 返回的是一个 dict_keys 视图对象，它没有 .sort() 方法。
# 只有列表才有 .sort() 方法。


# 📌 什么时候用 sorted()，什么时候用 .sort()？
# 方法	                 适用对象	                                       是否修改原对象
# sorted()	            任何可迭代对象（列表、元组、字典键等）	           ❌ 不修改原对象，返回新列表
# .sort()	            列表	                                           ✅ 修改原列表本身，不返回新列表
# dict_keys 不是列表，所以只能用 sorted()。
```

```python
# 集合：不允许有重复
favorite_language = {
    "jen": "python",
    "sarah": "C",
    "edward": "rust",
    "phil": "python",
}

for language in favorite_language.values():
    print(language)
# 不管有没有重复的值，都打印出来。

print()
for language in set(favorite_language.values()):
    #用set()把字典视图转化为一个集合。集合自动删掉了重复的值。
    print(language)
```

```python
# 嵌套：
# 字典中嵌套列表，列表中有字典。甚至字典中有字典

alien_0 = {"color": "green", "points": 5}
alien_1 = {"color": "yellow", "points": 20}
alien_2 = {"color": "red", "points": 15}

aliens = [alien_0, alien_1, alien_2]
# 此时，列表内的元素不能用引号。否则就是创建了一个新的列表，而不是列表里嵌套了字典。

for alien in aliens:
    print(alien)
```
