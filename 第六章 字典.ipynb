```python
# 因为 Python 中所有“通过索引或键来获取元素”的操作，统一使用方括号 []。

# 🔍 方括号 [] 的作用
# 数据类型	       用法	                           说明
# 列表	          list[0]	                      通过整数索引获取元素
# 元组	          tuple[0]	                      通过整数索引获取元素
# 字符串	      str[0]	                      通过整数索引获取字符
# 字典	          dict["key"]	                  通过键（key）获取值
# 核心规则： 方括号 [] 是 Python 中“下标访问/键访问”的统一语法，不管底层是序列还是映射。

# . 是用来访问属性和方法的，不是用来访问元素的。

# 有序序列（列表、元组、字符串）有整数下标。
# 字典没有整数下标，但使用键（key）来访问元素。
# 它们都用 []，但括号里的内容不同：一个是数字，一个是键名。

# 集合是无序数列，不能用下标(index)


# 数据类型	      是否有序	       用什么访问	             访问方式
# 列表	         ✅ 有序	         整数索引（下标）	        list[0]
# 元组	         ✅ 有序	         整数索引（下标）	        tuple[0]
# 字符串	     ✅ 有序	         整数索引（下标）	        str[0]
# 字典	         ✅ 有序（3.7+）	 键（key）	                dict["key"]
# 集合	         ❌ 无序	           不支持 []	—
```

```python
# 用for 循环，同时遍历所有的键和值。
user = {
    "username": "efermi",
    "first_name": "enrico",
    "last_name": "fermi",
}

for key in user:
    print(key)  # 每次循环结果空行
    # 遍历键名,value 同样。这是单独获得所有的键，或者所有的value.

for key, value in user.items():
    print("\n", key, value)
# 同时遍历所有的键和值。
# 变量：key 和 value可以是任意的，比如k,v或者name,value等等，前者代表键key，后者代表值value.
# 用键和值 循环得到的，是解包。不同于用一个变量，从字典视图得到的是元组。
```

```python
# 字典的方法：
user = {
    "username": "efermi",
    "first_name": "enrico",
    "last_name": "fermi",
}
print(user.keys())
# 获得字典的键

# ---> dict_keys(['username', 'first_name', 'last_name'])
# 不是元组，也不是列表。它是一个 dict_keys 对象，是字典的视图（view）

print(user.values())
# --->dict_values(['efermi', 'enrico', 'fermi'])

print(user.items())
# --->dict_items([('username', 'efermi'), ('first_name', 'enrico'), ('last_name', 'fermi')])
```

```python
# 字典中套字典：
# 网站的用户有自己的用户名，每个用户名关联自己的名、姓、居住地。
# 打印用户的信息

users = {
    "aeinstein": {
        "first_name": "albert",
        "last_name": "einstein",
        "location": "princeton",
    },
    "mcurie": {
        "first_name": "marie",
        "last_name": "curie",
        "location": "paris",
    },
}

for username, user_info in users.items():
    # username='awinstein' , 'mcurie'.   user_info=两个键的值，是字典的形式。
    print(f"\nusername:{username}")
    # \n表示每次遍历都以空一行开始。
    full_name = f"{user_info['first_name']} {user_info['last_name']}"
    # user_info已经是字典的形式。从字典提取value的方式就是dict[key]
    location = user_info["location"]

    print(f"\tfull_name:{full_name.title()}")
    print(f"\tlocation is {location}")
```

```python
people = []
# 定义一些人并将他们添加到前述列表中
person = {
    "first_name": "eric",
    "last_name": "matthes",
    "age": 46,
    "city": "sitka",
}
people.append(person)

person = {
    "first_name": "lemmy",
    "last_name": "matthes",
    "age": 2,
    "city": "sitka",
}
people.append(person)

person = {
    "first_name": "willie",
    "last_name": "matthes",
    "age": 11,
    "city": "sitka",
}
people.append(person)

# 先创建空列表，然后每次用同一个变量名添加到字典中，变量名可以重复使用，简化了每一次添加，就要有一个变量名的复杂化。

# 先创建空列表，然后添加。是以目标为导向的编程思维。更容易扩展维护，比如再添加几个人的信息，直接追加就可以。

# 显示列表包含的每个字典中的信息
for person in people:
    name = f"{person['first_name'].title()} {person['last_name'].title()}"
    age = person["age"]
    city = person["city"].title()
    print(f"{name}, of {city}, is {age} years old.")

# 输出：
# Eric Matthes, of Sitka, is 46 years old.
# Lemmy Matthes, of Sitka, is 2 years old.
# Willie Matthes, of Sitka, is 11 years old.
```

```python
# 嵌套关系的处理： 字典嵌套列表，或列表嵌套字典：

# 不论是字典嵌套列表，还是列表嵌套字典，都是先遍历外层，拿到内层的列表或字典，
# 然后再遍历内层的列表或字典，得到列表的元素，或字典的键和值。
# 再看要求，分别操作这些元素，或键和值。

# 📌 两种嵌套结构的遍历模式
# 1. 字典嵌套列表（外层是字典，内层是列表）

data = {
    "张三": ["大连", "哈尔滨", "长春"],
    "李四": ["洛阳", "开封", "许昌"],
}

for name, cities in data.items():  # 外层：遍历字典
    print(f"{name} 喜欢的城市：")
    for city in cities:  # 内层：遍历列表
        print(f"  {city}")


# 2. 列表嵌套字典（外层是列表，内层是字典）

data = [
    {"name": "张三", "cities": ["大连", "哈尔滨", "长春"]},
    {"name": "李四", "cities": ["洛阳", "开封", "许昌"]},
]

for item in data:  # 外层：遍历列表
    name = item["name"]
    cities = item["cities"]
    print(f"{name} 喜欢的城市：")
    for city in cities:  # 内层：遍历列表
        print(f"  {city}")

# 📌 核心规律
# 外层结构	        遍历方式	        内层结构	遍历方式
# 字典	            .items()	            列表	for item in list
# 字典	            .items()	            字典	.items()
# 列表	            for item in list	    字典	.items()
# 列表	            for item in list	    列表	for sub in list

# 📌 一句话总结
# 无论嵌套结构如何，遍历逻辑都是“先外层，再内层”：外层拿到内层容器，内层再遍历元素或键值对。
# 关键是搞清楚每一层是什么类型，用对应的遍历方式。
```
