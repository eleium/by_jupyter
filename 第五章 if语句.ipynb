```python
# 使用if语句处理列表：

# 菜谱是request_topping,客人点单：而green peppers 现在用完了

request_toppings = ["mushrooms", "extra cheese", "green peppers"]
for i in request_toppings:
    # 从菜谱中点单
    if i == "green peppers":
        # 如果客人点到 green peppers,就说：
        print(f"Sorry,we are out of {i} right now.")
    else:
        print(f"adding {i}")


print("\nfinished making your pizza!")

# 执行顺序：
# 开始第一次循环： i='mushrooms',因为不等于green peppers,执行else:-->add mushrooms
# 开始第二次循环： i="extra cheese",因为不等于green peppers,执行else:-->add extra cheese
# 开始第三次循环： i="green peppers",因为等于green peppers,执行print--->Sorry ,we are out of green peppers right now.

# 循环结束，执行最后一句：finished making your pizza!
```

```python
# 练习：序数
# 序数表示顺序位置：如1st 2nd. 序数多以th结尾，1st,2nd,3rd例外。

# 1,在一个列表中，存储数字1-9
# 2，遍历这个数列
# 3，在循环中，使用一个 if -elif-else 结构，打印每个数字对应的序数。
# 4，输出内容应该是： "1st 2nd 3rd 4th 5th 6th 7th 8th 9th",每个序数都独占一行。

num = range(1, 10)
for i in num:
    if i == 1:
        print("1st")
    elif i == 2:
        print("2nd")
    elif i == 3:
        print("3rd")
    elif i == 4:
        print("4th")
    elif i == 5:
        print("5th")
    elif i == 6:
        print("6th")
    elif i == 7:
        print("7th")
    elif i == 8:
        print("8th")
    else:
        print("9th")
print("-" * 40)
# 简洁版：
numbers = range(1, 10)
for number in numbers:
    if number == 1:
        print("1st")
    elif number == 2:
        print("2nd")
    elif number == 3:
        print("3rd")
    else:
        print(f"{number}th")
# 先单独表达后缀特殊的，然后用f-string f字符串方法表达通用的。
```

```python
# 另外一种用后缀的方法：

num = range(1, 10)

for i in num:
    if i == 1:
        suffix = "st"
        # suffix 在循环体内定义也来得及。可以。
    elif i == 2:
        suffix = "nd"
    elif i == 3:
        suffix = "rd"
    else:
        suffix = "th"
    print(f"{i}{suffix}")
```
